```python
import pandas as pd

COLUMNS_MAPPING = {
    'Timestamp': 'datetime64[ns]',
    'NIP': 'string'
    'Company Code - Name': 'string',
    'Department': 'string',
    'Request Type': 'string',
    'Promo Type': 'string',
    'Respon Requester': 'string',
    'Timestamp Requester': 'datetime64[ns]',
    'Name Requester': 'string',
    'Respon Approver': 'string',
    'Timestamp Approver': 'datetime64[ns]',
    'Name Approver': 'string',
    'Processed By': 'string',
    'Process Status': 'string',
    'Taken Date': 'datetime64[ns]',
    'Processed Date': 'datetime64[ns]',
    'Feedback Status': 'string',
    'Request Type Group': 'string'
}

def load_data():
    data = pd.DataFrame()
    #TODO: Adjust later
    excel = pd.ExcelFile('../data/SPREADSHEET_NAME_HERE.xlsx', engine='openpyxl')

    columns = list(COLUMNS_MAPPING.keys())

    for sheet in excel.sheet_names[:-3]:
        df = pd.read_excel(excel, sheet, skiprows=1)
        df['Request Type Group'] = sheet    
        df.columns = df.columns.str.strip()
        
        for column, dtype in COLUMNS_MAPPING.items():
            if column in df.columns:
                df[column] = df[column].astype(dtype, errors='ignore')
        
        
        data = pd.concat([data.copy(), df[columns]], ignore_index=True)
    
    return data
```
